```python
import glob

fnames=glob.glob("output_*")
fnames_int=np.array([fname.replace('output_','').replace('.asc','') for fname in fnames])
fnames_tsnaps=np.sort(fnames_int.astype(int))
```

```python
def init():
    line.set_data([], [])
    return (line,)


def animate(i):
        
    u_wave = np.loadtxt(f"output_{i}.asc")
    
    line.set_data(u_wave[:,0], u_wave[:,1])
    return (line,)
```

```python
fig, ax = plt.subplots(figsize=[15, 5])

ax.set_xlim(( 0, 1))
ax.set_ylim((-0.5, 1.5))
ax.set_xlabel('Position x')
ax.set_ylabel('Solution')

line, = ax.plot([], [], lw=2, label = 'u (x,t)')

ax.legend()
```

```python
from matplotlib import animation, rc
from IPython.display import HTML

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=fnames_tsnaps,
                               interval=100,
                               blit=True)
```
